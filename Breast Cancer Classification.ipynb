## BREAST CANCER CLASSIFICATION

### IMPORTING THE DATASET

```python
import pandas as pd 
import numpy as np 
import matplotlib.pylab as plt 
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

```python
cancer.keys()
```

```python
print(cancer ["DESCR"])
```

```python
print(cancer ["target"])
```

```python
print(cancer ["target_names"])
```

```python
print(cancer ["feature_names"])
```

```python
cancer["data"].shape
```

```python
df_cancer = pd.DataFrame(np.c_[cancer["data"], cancer["target"]], columns=np.append(cancer["feature_names"],
                                                                                   ["target"]))

df_cancer.head()
```

```python
df_cancer.tail()
```

### VISUALIZATING THE DATA

```python
sns.pairplot(df_cancer, hue="target", vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'])
```

```python
sns.countplot(df_cancer["target"])
```

```python
sns.scatterplot(x = "mean area", y="mean smoothness", hue="target",
               data=df_cancer)
```

```python
plt.figure(figsize= (20, 18))
sns.heatmap(df_cancer.corr(), annot=True)
```

### MODEL TRAINING 

```python
X = df_cancer.drop(["target"], axis=1)
X
```

```python
y = df_cancer["target"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train
```

```python
y_test
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
svc_model.fit(X_train, y_train)
```

### EVALUATING THE MODEL

```python
y_pred = svc_model.predict(X_test)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm 
```

```python
sns.heatmap(cm, annot=True)
```

### IMPROVING THE MODEL

```python
min_train = X_train.min()
```

```python
range_train = (X_train - min_train).max()
```

```python
X_train_scaled = (X_train - min_train) / range_train
```

```python
sns.scatterplot(x = "mean area", y="mean smoothness", hue="target",
               data=df_cancer)
```

```python
sns.scatterplot(x = X_train_scaled["mean area"], y=X_train_scaled["mean smoothness"], hue="target",
               data=df_cancer)
```

```python
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test) / range_test
#X_test_scaled
```

```python
svc_model.fit(X_train_scaled, y_train)
```

```python
y_pred = svc_model.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

### IMPROVING THE MODEL  - PART 2

```python
param_grid = {"C": [0.1, 1, 10, 100], "gamma": [1, 0.1, 0.01, 0.001], "kernel": ['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose=4)
grid.fit(X_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
grid_predictions = grid.predict(X_test_scaled)
grid_predictions
```

```python
cm = confusion_matrix(y_test, grid_predictions)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, grid_predictions))
```
